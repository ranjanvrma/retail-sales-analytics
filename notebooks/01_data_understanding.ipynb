# Questions

Set 1

```python
# Which product categories generate the highest sales?

df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(1)
```

```python
# Which sub-categories generate the highest sales?

df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(1)
```

```python
# Which customer segment contributes the most sales?
df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).head(1)
```

```python
# Which states generate the highest sales?
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(1)
```

```python
# Which shipping mode is used the most?
df['Ship Mode'].value_counts().head(1)
```

Set 2

```python
# Which Category generated the highest sales in each Region?
sales = df.groupby(['Region', 'Category'])['Sales'].sum()

sales.loc[sales.groupby(level=0).idxmax()]
```

```python
# Which State has the highest sales within each Category?
df.groupby(['Category', 'State'])['Sales'].sum()
```

```python
# Which Ship Mode is most commonly used by each Segment?
df.groupby(['Segment', 'Ship Mode']).size()
```

```python
# Which Sub-Category sells the most in each State?
sales = df.groupby(['State', 'Sub-Category'])['Sales'].sum()

sales.loc[sales.groupby(level=0).idxmax()]
```

```python
# Which Customer purchased the highest total sales?
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending = False)
```

Set 3

```python
# For each Category, show: Total Sales, Average Sales, Highest Sale, Lowest Sale
df.groupby('Category')['Sales'].agg(
    Total_Sales = 'sum',
    Average_Sales = 'mean',
    Highest_Sale = 'max',
    Lowest_Sale = 'min'
)
```

```python
# For each Region, show: Total Sales, Average Sales, Number of Transactions
df.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
```

```python
# For each Customer Segment, show: Total Sales, Highest Sale, Lowest Sale, Average Sales
df.groupby('Segment')['Sales'].agg(['sum', 'max', 'min', 'mean'])
```

```python
# For each Ship Mode, show: Number of Orders, Total Sales, Average Sales
df.groupby('Ship Mode')['Sales'].agg(['count', 'sum', 'mean'])
```

```python
# For each State, show: Number of Customers, Total Sales
df.groupby('State')['Sales'].agg(['nunique', 'sum'])
```

Set 4

```python
# Show all rows where: State == 'California'
df[df['State'] == 'California']
```

```python
# Show all rows where: Category == 'Technology'
df.query("Category == 'Technology'")
```

```python
# Show all rows where: Sales > 1000
df[df['Sales'] > 1000]
```

```python
# Show all Technology products with Sales greater than 1000.
df[(df['Category'] == 'Technology') & (df['Sales'] > 1000)]
```

```python
# Show all orders from: California, Technology and Sales > 1000
df.query("State == 'California' and Category == 'Technology' and Sales > 1000")
```

Set 5

```python
# Create a table where: Rows → Region, Values → Total Sales
table = df.pivot_table(index = 'Region', values = 'Sales', aggfunc= 'sum')
table
```

```python
# Create a table where: Rows → Category, Columns → Region, Values → Sales (sum)
table = df.pivot_table(index = 'Category', columns = 'Region', values = 'Sales', aggfunc = 'sum')
table
```

```python
# Rows: Segment, Columns: Ship Mode, Values: Sales (sum)
table = df.pivot_table(index = 'Segment', columns = 'Ship Mode', values = 'Sales', aggfunc = 'sum')
table
```

```python
# Rows: State, Values: Average Sales
table = df.pivot_table(index = 'State', values = 'Sales', aggfunc = 'mean')
table
```

```python
# Rows: Category, Columns: Segment, Values: Sales (sum), Margins: Yes (Grand Total)
table = df.pivot_table(index = 'Category', columns = 'Segment', values = 'Sales', aggfunc = 'sum', margins = True)
table
```
