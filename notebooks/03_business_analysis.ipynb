```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/superstore_sales.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

## Sales Analysis

```python
# What is the total revenue?
Total_Sales = df['Sales'].sum()
print(Total_Sales)
```

```python
# What is the average sale amount?
Average_Sales = df['Sales'].mean()
print(Average_Sales)
```

```python
# Which category generates the highest sales?
df.groupby('Category')['Sales'].sum().sort_values(ascending = True)
```

```python
# Which sub-category generates the highest sales?
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending = False).head()
```

```python
# Top 10 products by sales.
df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(10)
```

```python
# Bottom 10 products by sales.
df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).tail(10)
```

```python
# What percentage of total sales comes from each category?
df.groupby('Category')['Sales'].sum() / df['Sales'].sum() * 100
```

```python
# What percentage of total sales comes from each region?
df.groupby('Region')['Sales'].sum() / df['Sales'].sum() * 100
```

## Customer Analysis

```python
# How many unique customers are there?
df['Customer ID'].nunique()
```

```python
# Top 10 customers by sales.
df.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending = False).head(10)
```

```python
# Average spending per customer.
df.groupby(['Customer ID', 'Customer Name'])['Sales'].mean()
```

```python
# Which customer segment contributes the most revenue?
df.groupby('Segment')['Sales'].sum().sort_values(ascending = False)
```

```python
# Which customers placed the most orders?
df[['Customer ID', 'Customer Name']].value_counts().head()
```

```python
# Find repeat customers (customers with more than one order).
df['Customer ID'].value_counts()[lambda x: x > 1]
```

## Product Analysis

```python
# Average sales for each category.
df.groupby('Category')['Sales'].mean()
```

```python
# Average sales for each sub-category.
df.groupby('Sub-Category')['Sales'].mean()
```

```python
# Best-selling product in each category.
product_sales = df.groupby(['Category', 'Product Name'])['Sales'].sum().sort_values(ascending = False)
product_sales.loc[product_sales.groupby(level = 0).idxmax()]
```

```python
# Top 5 products in Technology.
tech = df[df['Category'] == 'Technology']
tech.groupby( 'Sub-Category')['Sales'].sum().sort_values(ascending = False)
```

## Geographic Analysis

```python
# Top 10 cities by sales.
df.groupby('City')['Sales'].sum().sort_values(ascending = False).head(10)
```

```python
# Top 10 states by sales.
df.groupby('State')['Sales'].sum().sort_values(ascending = False).head(10)
```

```python
# Which region generates the highest revenue?
df.groupby('Region')['Sales'].sum().sort_values(ascending = False)
```

```python
# Best-selling category in every region.
best_selling = df.groupby(['Region', 'Category'])['Sales'].sum()
best_selling.loc[best_selling.groupby(level = 0).idxmax()]
```

```python
# Average sales in each state.
df.groupby('State')['Sales'].mean().head()
```

## Time Analysis

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%d/%m/%Y')
```

```python
# Year-wise sales.
df.groupby(df['Order Date'].rename('Year').dt.year)['Sales'].sum()
```

```python
# Month-wise sales.
monthly_sales = df.groupby(df['Order Date'].rename('Month').dt.month)['Sales'].sum()
monthly_sales.index = pd.to_datetime(monthly_sales.index, format = '%m').month_name()
monthly_sales
```

```python
# Quarter-wise sales.
df.groupby(df['Order Date'].rename('Quarter').dt.quarter)['Sales'].sum()
```

```python
# Highest sales month.
monthly_sales = df.groupby(df['Order Date'].rename('Month').dt.month)['Sales'].sum().sort_values(ascending = False)
monthly_sales.index = pd.to_datetime(monthly_sales.index, format = '%m').month_name()
monthly_sales.head(1)
```

```python
# Highest sales year.
df.groupby(df['Order Date'].dt.year)['Sales'].sum().sort_values(ascending = False)
```
