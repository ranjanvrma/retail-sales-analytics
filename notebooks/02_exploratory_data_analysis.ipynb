## Sales Performance

```python
# Which region generates the highest sales?
df.groupby('Region')['Sales'].sum().sort_values(ascending = False)
```

```python
# Which state generates the highest sales?
df.groupby('State')['Sales'].sum().sort_values(ascending = False).head()
```

```python
# Which category generates the highest sales?
df.groupby('Category')['Sales'].sum().sort_values(ascending = False)
```

```python
# Which sub-category generates the highest sales?
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending = False).head()
```

```python
# Which products generate the highest sales?
df.groupby(['Product ID', 'Product Name'])['Sales'].sum().sort_values(ascending = False).head()
```

## Customer Analysis

```python
# Which customer has spent the most?
df.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending = False).head()
```

```python
# Which customer segment contributes the most revenue?
df.groupby('Segment')['Sales'].sum().sort_values(ascending = False)
```

```python
# Top 10 customers by sales.
df.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending = False).head(10)
```

## Shipping

```python
# Which shipping mode is used the most?
df.groupby('Ship Mode')['Order ID'].count().sort_values(ascending = False)
```

```python
# Which shipping mode generates the highest sales?
df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending = False)
```

## Geography

```python
# Top 10 cities by sales.
df.groupby('City')['Sales'].sum().sort_values(ascending = False).head(10)
```

```python
# Sales distribution by region.
df.groupby('Region')['Sales'].sum().sort_values(ascending = False)
```

## Time Analysis

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format = '%d/%m/%Y')
```

```python
# Sales by Year
df.groupby(df['Order Date'].dt.year)['Sales'].sum()
```

```python
# Sales by Month
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

monthly_sales.index = pd.to_datetime(monthly_sales.index, format = '%m').month_name()
monthly_sales
```

```python
# Which month has the highest sales?
df.groupby(df['Order Date'].dt.month_name())['Sales'].sum().sort_values(ascending = False).head()
```

```python
# Which year had the highest sales?
df.groupby(df['Order Date'].dt.year)['Sales'].sum().sort_values(ascending = False)
```
